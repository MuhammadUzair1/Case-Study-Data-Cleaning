# Client Data

```python
data2 = data.copy()
```

```python
filtered_data = data2[data2['job'].str.contains(r'\.', na=False)]
filtered_data.head()
```

# Chaniging '.' to "_"

```python
data2['job'] = data2['job'].str.replace('.', '_')
data2
```

# Change "." to "_" and "unknown" to np.NaN   in Column education

```python
filtered_data2 = data2[data2['education'].str.contains(r'\.', na=False)]
filtered_data2.head()
```

```python
data2['education'] = data2['education'].str.replace('.', '_')
data2.head()
```

```python
data2['education'] = data2['education'].replace('unknown', np.NaN) 
data2.head(10)
```

```python
data2['credit_default'] = data2['credit_default'].map({'yes': '1', 'no': '0'}).fillna(0)
data2.tail() 
# data.head(10) 
```

```python
data2['mortgage'] = data2['mortgage'].map({'yes': '1', 'no': '0'}).fillna(0)
data2.head() 
# data.head(10) 
```

```python
# Columns
columns_to_extract = ['client_id', 'age', 'job', 'marital', 'education', 'credit_default', 'mortgage']
# Check which columns exist in the file to avoid errors
existing_columns = [col for col in columns_to_extract if col in data2.columns]

if existing_columns:
    extracted_df = data2[existing_columns] 
else:
    extracted_df = pd.DataFrame(columns=columns_to_extract)

extracted_df.to_csv('client.csv', index=False)

print("Data has been saved to client.csv")
```

# Campaign

```python
# Convert to boolean data type:
# 1 if "success", otherwise 0.
data2['previous_outcome'] = data2['previous_outcome'].apply(lambda x: 1 if str(x).lower() == 'success' else 0)
```

```python
data2.tail(10)
```

```python
# Convert to boolean data type:
# 1 if "success", otherwise 0.
data2['campaign_outcome'] = data2['campaign_outcome'].apply(lambda x: 1 if str(x).lower() == 'success' else 0) 
```

```python
data2.tail()
```

```python
from datetime import datetime
# Add a new 'year' column with a default value (e.g., 2024)
data2['year'] = 2024

# Define a month name to number mapping
month_map = {
    'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6,
    'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12
}

# Replace month names with their corresponding numbers
data2['month'] = data2['month'].str.strip().str.lower().map(month_map)

# Create 'last_contact_date' column using datetime
data2['last_contact_date'] = data2.apply(
    lambda row: datetime(row['year'], row['month'], row['day']).strftime('%Y-%m-%d'), axis=1
)
```

```python
data2.head()
```

```python
# Columns
columns_to_extract = ['client_id', 'number_contacts', 'contact_duration', 'previous_campaign_contacts', 'previous_outcome', 'campaign_outcome', 'last_contact_date']

# Extract the selected columns and save to a new file
extracted_df = data2[columns_to_extract]

extracted_df.to_csv('campaign.csv', index=False)


print("Data has been saved to campaign.csv")
```
